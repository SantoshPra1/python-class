```python
#24 Remove all the special characters and numbers from the following string
#text = '%p34@y!*-*!t68h#&on404'

s = '%p34@y!*-*!t68h#&on404'
s1 =s.replace('%','')
s2 =s1.replace('@','')
s3 =s2.replace('!','')
s4 =s3.replace('-','')
s5 =s4.replace('&','')
s6 =s5.replace('#','')
s7 =s6.replace('*','')

print(s7)
```
